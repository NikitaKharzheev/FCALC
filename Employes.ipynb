```python
import fcalc
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

# Binarized Data

```python
df = pd.read_csv('Employee.csv')

df.head()
```

```python
df.dropna(inplace=True)
df.info()
```

```python
df.nunique ()
```

```python
from sklearn.utils import shuffle
df = shuffle(df)
df=df.iloc[1:500]
df = df.sample(frac=1).reset_index(drop=True)
df
```

```python
df.nunique ()
```

```python
print (df['Age'].min())
print (df['Age'].max())
```

```python
bin_data = dict()
bin_data['Age22_28'] = (df['Age']<=28).astype(int)
bin_data['Age28_35'] = ((df['Age']>28) & (df['Age']<=35)).astype(int)
bin_data['Age35_41'] = (df['Age']>35).astype(int)
```

```python
print (df['JoiningYear'].min())
print (df['JoiningYear'].max())
```

```python
bin_data['late_year'] = (df['JoiningYear']<=2015).astype(int)
bin_data['not_late_year'] = (df['JoiningYear']>2015).astype(int)
```

```python
print (df['ExperienceInCurrentDomain'].min())
print (df['ExperienceInCurrentDomain'].max())
```

```python
bin_data['small_exp'] = (df['ExperienceInCurrentDomain']<=3).astype(int)
bin_data['big_exp'] = (df['ExperienceInCurrentDomain']>3).astype(int)
```

```python
df_bin = pd.DataFrame(bin_data)
df_bin = df_bin.sample(frac=1).reset_index(drop=True)
df_bin
```

```python
df_new=pd.concat([df_bin, df], sort=False, axis=1)
df_new
```

```python
df_new['LeaveOrNot'] = [x == 1 for x in df_new['LeaveOrNot']]
df_new.sample(10)
```

```python
df_new_new=df_new.iloc[:,:-1]
X = pd.get_dummies(df_new_new).astype(bool)
y = df_new['LeaveOrNot']
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
bin_cls = fcalc.classifier.BinarizedBinaryClassifier(X_train.values, y_train.to_numpy(), method="standard-support")
```

```python
bin_cls.predict(X_test.values)
```

```python
from sklearn.metrics import accuracy_score

print("Accuracy:", accuracy_score(y_test, bin_cls.predictions))
```

# Pattern structure

```python
df1 = pd.read_csv('Employee.csv')
from sklearn.utils import shuffle
df1 = shuffle(df1)
df1=df1.iloc[1:500]
df1 = df1.sample(frac=1).reset_index(drop=True)
df1
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df1.loc[:, "Education"] = le.fit_transform(df1["Education"])
df1.loc[:, "City"] = le.fit_transform(df1["City"])
df1.loc[:, "PaymentTier"] = le.fit_transform(df1["PaymentTier"])
df1.loc[:, "Gender"] = le.fit_transform(df1["Gender"])
df1.loc[:, "EverBenched"] = le.fit_transform(df1["EverBenched"])
```

```python
df1['LeaveOrNot'] = [x == 1 for x in df1['LeaveOrNot']]
df1
```

```python
X = df1.iloc[:,:-1]
y = df1['LeaveOrNot']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
pat_cls = fcalc.classifier.PatternBinaryClassifier(X_train.values, y_train.to_numpy(), 
                                                   categorical=np.array([0,2,3,5,6]))
```

```python
pat_cls.predict(X_test.values)
```

```python
from sklearn.metrics import accuracy_score, f1_score
print("accuracy:",round(accuracy_score(y_test, pat_cls.predictions),4))
```

```python
from sklearn.metrics import accuracy_score, classification_report, f1_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
def try_models(model):
    model.fit(X_train, y_train)
    
    y_preds = model.predict(X_test)
    
    acc_score = round(accuracy_score(y_test, y_preds),4)
    
    return (f'Accuracy Score of {model}: {acc_score}')
```

```python
try_models(KNeighborsClassifier())
```

```python
try_models(LogisticRegression())
```

```python
try_models(DecisionTreeClassifier())
```

```python
try_models(RandomForestClassifier(random_state=42))
```
